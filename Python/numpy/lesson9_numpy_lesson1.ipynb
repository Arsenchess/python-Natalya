# Векторная алгебра .
Скаляр - это одно простое число
Физические величины:
1. Скалярные - величины, которые имеют только числовое значение (модуль) и не имеют направлениея(длина, температура(тела на воздухе), время, обьем, площадь, масса,высота, плотность тела).
2. Векторные - величины, которые кроме числового значения (модуль), имеют и направления. Про векторную величину можно всегда сказать куда она направлена (сила, скорость, перемещение, ускорение).

Множество натуральных чисел:

## Множество - это совокупность объектов, объединенных общим свойством.
Объекты из которого состоят множества называются элементами этого множества.
Если 'a' элемент множества 'A', то записывают:
a ∈ A (a принадлежит  A).
Если нет - a  ∉  A
## Способы создания множеств:
1) перечисление
2) описание (определяем свойства их элементов)
### Диаграмма Венна

∅ - это знак пустого множества.
Пустое множество это множество, где нет элементов.
0 ∈ N - Элемент 0 принадлежит множеству N.
 ∈  - этот знак показывает отношение между элементом и множеством.
⊂ -  показывает отношение между множеством и множеством.
N ⊂  Z (Множество N (натуральные числа) входит во множство Z (целых чисел).
